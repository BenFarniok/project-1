```python
import os
import json
import pandas as pd
```

```python
filepath = os.path.join("resources", "2018_fire_data_cleaned.json")
f=open(filepath)
late_data=json.load(f) 

with open(filepath) as jsonfile:
    fire_json = json.load(jsonfile)
print(json.dumps(fire_json, indent=4))
```

```python
print(len(fire_json))
i=0
names=[]
acreage=[]
latitude=[]
longitude = []
start = []
extinguished = []
for fire in fire_json:
    #fire_json[i]['Name'])
    names.append(fire_json[i]['Name'])
    acreage.append(fire_json[i]['AcresBurned'])
    latitude.append(fire_json[i]['Latitude'])
    longitude.append(fire_json[i]['Longitude'])
    start.append(fire_json[i]['Started'])
    extinguished.append(fire_json[i]['Extinguished'])
    i=i+1
```
