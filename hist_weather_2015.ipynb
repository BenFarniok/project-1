```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import os
import json
# Import API key
#from api_keys import gkey
file_to_load = 'outputs/2015_fires'
pd.set_option('mode.chained_assignment', None)
# Read Purchasing File and store into Pandas data frame
fire_df = pd.read_csv(file_to_load)
```

```python
nocoords = fire_df[fire_df['Latitude'] == 0.00000 ].index
fire_df.drop(nocoords, inplace= True)
fire_df.count()
```

```python
latitude = []
longitude = []
datelist=[]
temp = []
windspeed= []
precip = []
humidity = []
visi = []
pressure = []
cloudcover = []
windgust = []
uvindex = []
location = ''
date = ''
```

```python
base_url = "http://api.worldweatheronline.com/premium/v1/past-weather.ashx"
api_key='YOUR KEY HERE'
#location = '35.924360,-119.940830'

for index, row in fire_df.iterrows():
    lat = row["Latitude"]
    lng = row["Longitude"]
    location = f"{lat},{lng}"
    date = row['Fire Started']
    call = f'http://api.worldweatheronline.com/premium/v1/past-weather.ashx?key={api_key}&q={location}&format=json&date={date}&endate={date}&tp=24'
    response = requests.get(call).json()
    try:      
        datelist.append((response['data']['weather'][0]['date']))
        temp.append(response['data']['weather'][0]['maxtempF'])
        windspeed.append(response['data']['weather'][0]['hourly'][0]['windspeedMiles'])
        precip.append(response['data']['weather'][0]['hourly'][0]['precipMM'])
        humidity.append(response['data']['weather'][0]['hourly'][0]['humidity'])
        visi.append(response['data']['weather'][0]['hourly'][0]['visibilityMiles'])
        pressure.append(response['data']['weather'][0]['hourly'][0]['pressureInches'])
        cloudcover.append(response['data']['weather'][0]['hourly'][0]['cloudcover'])
        windgust.append(response['data']['weather'][0]['hourly'][0]['WindGustMiles'])
        uvindex.append(response['data']['weather'][0]['hourly'][0]['uvIndex'])
        latitude.append(row['Latitude'])
        longitude.append(row['Longitude'])
 #   hotel_df.loc[index, "Hotel Name"] = response["results"][0]["name"]
    except:
        print(f"couldn't find information for {index}" )
#print(response)    
  #  hist_weather_data_2010 = retrieve_hist_data(api_key,
                               # location,
                               # date,
                               # end_date,
                               # frequency,
                               # location_label = False,
                               # export_csv = True,
                               # store_df = True)
```

```python
data_df = pd.DataFrame({'Latitude': latitude, 
            'Longitude': longitude, 
            'Precipitation(MM)':precip, 
           'Humidity':humidity,
           'Temperature':temp, 
           'Humidity': humidity,
           'Wind Speed': windspeed,
            'Wind Gust': windgust,
            'Temperature':temp,
           'Cloud Cover':cloudcover,
            'Date':datelist})
data_df
```

```python
data_df.to_csv("outputs/fireweather2015.csv" , index = True)
```
