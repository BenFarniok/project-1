```python
import os
import json
import pandas as pd
```

```python
filepath = os.path.join("resources", "2015_fire_data.json")
f=open(filepath)
late_data=json.load(f) 

with open(filepath) as jsonfile:
    fire_json = json.load(jsonfile)
print(json.dumps(fire_json, indent=4))
```

```python
print(len(fire_json))
i=0
names=[]
acreage=[]
latitude=[]
longitude = []
start = []
extinguished = []
counties=[]
for fire in fire_json:
    #fire_json[i]['Name'])
    names.append(fire_json[i]['Name'])
    acreage.append(fire_json[i]['AcresBurnedDisplay'])
    latitude.append(fire_json[i]['Latitude'])
    longitude.append(fire_json[i]['Longitude'])
    start.append(fire_json[i]['StartedDate'])
    extinguished.append(fire_json[i]['UpdatedDate'])
    counties.append(fire_json[i]['CountiesList'])
    i=i+1
```

```python
fire_df = pd.DataFrame({'Name':names, 
                             'Acres Burned':acreage, 
                             'Latitude':latitude, 
                             'Longitude':longitude, 
                             "Fire Started":start, 
                             'Fire Extinguished':extinguished,
                        'Counties':counties})

fire_df
```

```python
#fire_df["Date Started"]= fire_df['Fire Started'].str.split("T", n = 1, expand = True)
fire_df["Fire Extinguished"]= fire_df['Fire Extinguished'].str.split(" ", n = 1, expand = True)

fire_df
```

```python
fire_df_clean= fire_df[['Name','Counties','Acres Burned', 'Latitude', 'Longitude', 'Fire Started', 'Fire Extinguished']]
fire_df_clean
```
